<h1> Graphing Distributions </h1> Since we'll be looking at the distrubution between gender and the gendre of the author. The below, gives us a sense of those two variables. I had the help of CHATGPT to deal with altair as I usually use mathplot and not as familar with this library.

```python
genre_count = df['genre'].value_counts().reset_index()
gender_count = df['author_gender'].value_counts().reset_index()
genre_count.columns = ["Genre", "Count"]
gender_count.columns = ['Gender','Count']

```

```python



genre_chart = alt.Chart(genre_count).mark_bar().encode(
    x=alt.X('Count:Q', title='Number of Books'),
    y=alt.Y('Genre:N', sort='-x', title='Genre')
).properties(
    title='Genre Distribution in Top 500 Novels'
)

genre_chart
```

```python

gender_chart = alt.Chart(gender_count).mark_bar().encode(
    x=alt.X('Count:Q', title='Number of Authors'),
    y=alt.Y('Gender:N', sort='-x', title='Gender')
).properties(
    title='Gender Distribution of Authors in Top 500 Novels'
)

gender_chart
```

```python



gender_genre_distribution = pd.crosstab(df['author_gender'], df['genre'])


gender_genre_melted = gender_genre_distribution.reset_index().melt(id_vars="author_gender", var_name="Genre", value_name="Count")


stacked_bar = alt.Chart(gender_genre_melted).mark_bar().encode(
    x='author_gender:N',  
    y='Count:Q',           
    color='Genre:N',       
    tooltip=['author_gender', 'Genre', 'Count']  
).properties(
    title="Genre Distribution Across Genders"
)

stacked_bar

```
